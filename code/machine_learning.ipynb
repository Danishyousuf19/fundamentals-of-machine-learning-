```python
# label encoder
from sklearn import preprocessing
header= ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
encoder=preprocessing.LabelEncoder()
encoder.fit(header)
# testing
test=['class','pres']
encoded_values=encoder.transform(test)
print("\nlabels=",test)
print("encoded = ",list(encoded_values))
encoded_values=[3,6,4,1]
decoded_list=encoder.inverse_transform(encoded_values)
print("\nencoded=",encoded_values)
print("decoded=",list(decoded_list))
```

```python
# univariate selection
from sklearn.feature_selection import SelectKBest , chi2
path =r"C:\Users\DELL\Downloads\INTERN\pima-indians-diabetes.csv"
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = read_csv(path, names=names)
array=data.values
x=array[:,0:8]
y=array[:,8]
test =SelectKBest(score_func=chi2,k=4)
fit=test.fit(x,y)
set_printoptions(precision=2)
print(fit.scores_)
featured=fit.transform(x)
print("\nfeatured\n",featured[0:4])
```

```python
# recursive feature elemination
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
path = r"C:\Users\DELL\Downloads\INTERN\pima-indians-diabetes.csv"
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = read_csv(path, names=names)
array = data.values
x = array[:, 0:8]
y = array[:, 8]
model =LogisticRegression()
rfe=RFE(model,n_features_to_select=3)
fit=rfe.fit(x,y)
print("Number of Features: %d" % fit.n_features_)
print("Selected Features: %s" % fit.support_)
print("Feature Ranking: %s" % fit.ranking_)
```

```python
# principle component  analysis 
from sklearn.decomposition import PCA
from numpy import set_printoptions
path = r"C:\Users\DELL\Downloads\INTERN\pima-indians-diabetes.csv"
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = read_csv(path, names=names)
array = data.values
x = array[:, 0:8]
y = array[:, 8]
pca=PCA(n_components=3)
set_printoptions(precision=2)
fit=pca.fit(x)
print("Explained Variance: %s"% fit.explained_variance_ratio_)
print(fit.components_)
```

```python
# feature importance
from sklearn.ensemble import ExtraTreesClassifier
path = r"C:\Users\DELL\Downloads\INTERN\pima-indians-diabetes.csv"
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = read_csv(path, names=names)
array = data.values
x = array[:, 0:8]
y = array[:, 8]
model=ExtraTreesClassifier()
model.fit(x,y)
print(model.feature_importances_)
```
